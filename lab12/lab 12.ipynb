```python
%matplotlib inline
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph() 
```

```python
G.add_nodes_from('SciCompPy')
for (a, b) in [('S', 'c'), ('S', 'i'), ('C', 'o'), ('C', 'm'),
               ('C', 'p'), ('P', 'y'), ('S', 'C'), ('C', 'P'),
               ('P', 'S')]:
    G.add_edge(a, b)
```

```python
nx.draw(G, with_labels=True, font_weight='bold')
plt.show()
```
