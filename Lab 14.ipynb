```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
def approximate_pi(N):
    points = np.random.uniform(0, 1, (N, 2))
    return (points[(np.sum(points**2, axis=1) <= 1)].shape[0])/points.shape[0] * 4
```

```python
for i in range(1, 7):
    print('Approximate value of Pi using {} points: {}'.format(10**i, approximate_pi(10**i)))
```

```python
def approximate_pi_plot(N):
    plt.figure(figsize=(12,9))
    points = np.random.uniform(0, 1, (N, 2))
    pi = (points[(np.sum(points**2, axis=1) <= 1)].shape[0])/points.shape[0] * 4
    colors = [(0, 1, 0) if cond else (1, 0, 0) for cond in np.less_equal(points[:, 0]**2 + points[:, 1]**2, 1)]
    plt.scatter(points[:, 0], points[:, 1], c = colors, label=colors)
    plt.grid()
    plt.title('Approximate value of Pi using {} points: {}'.format(N, pi))
    green_patch = mpatches.Patch(color=(0, 1, 0), label='Points inside circle')
    red_patch = mpatches.Patch(color=(1, 0, 0), label='Points outside circle')
    plt.legend(handles=[green_patch, red_patch], bbox_to_anchor=(1, 0.5))
    plt.show()
```

```python
a = approximate_pi_plot(10**4)
```
