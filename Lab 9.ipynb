```python
# make matplotlib graphs included in the notebook, next to the code.
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
from typing import Tuple
```

## Exercise 9.1

```python
def plot(a: float = 0, b: float = 10, number: int = 1000,
         figsize: Tuple[float, float] = (12, 9)) -> None:
    '''
    Plot functions sin(x), cos(x), and exp(-x) in a range [a, b] with colors
    red, green, and blue, respectively and linestyles solid, dashed, and dotted,
    respectively. Legend is provided.
    
    Args:
        a - start point of the interval.
        b - end point of the interval.
        number - number of samples to generate interval [a, b].
        figsize - size of the figure.
    Returns:
        None.
    '''
    plt.figure(figsize=(12, 9))
    x = np.linspace(a, b, number) 
    plt.plot(x, np.sin(x), linestyle = 'solid', color = (1, 0, 0), label = 'sin(x)')
    plt.plot(x, np.cos(x), linestyle = 'dashed', color = (0, 1, 0), label = 'cos(x)')
    plt.plot(x, np.exp(-x), linestyle = 'dotted', color = (0, 0, 1), label = 'exp(-x)')
    plt.legend(loc = "best") # place the legend at best location
    plt.grid() # show grid
    plt.show()
```

```python
plot()
```

## Exercise 9.2

```python
def scatter(a: float = 0, b: float = 0, number: int  = 100,
            figsize: Tuple[float, float] = (12, 9)) -> None:
    '''
    Generate number of random points in a square [a, b] x [a, b]. Points are green
    if the distance from (a, a) is less then b, they are red otherwise. 
    The area of a point (x, y) is 200 times x + y.
    Args:
        a - start point of the interval.
        b - end point of the interval.
        number - number of points to generate.
        figsize - size of the figure.
    '''
    plt.figure(figsize=figsize)
    np.random.seed(23) # Fixing random state for reproducibility
    x = np.random.rand(number)
    y = np.random.rand(number)
    area = 200 * (x + y)
    colors = [(0, 1, 0) if cond else (1, 0, 0) for cond in np.less(x**2 + y**2, 1)]
    plt.scatter(x, y, s = area, c = colors)
    plt.grid() # show grid
    plt.show()
```

```python
scatter()
```
